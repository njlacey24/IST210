    Our group has decided to break down this lab into seven entities. We based this off of the idea giving separate entities for both fish and pets and associated their separate habitats (enclosures for pets and aquariums for fish) by having a foreign key present within each habitat entity that will show the type of animal that is supposed to go to each habitat. We further broke down the money involvement in the store by having Purchases be related to customer purchases and Expense Report being involved with the food that the store buys for the animals. This was both a requirement of the assignment but also an important aspect for a business to incorporate. Realistically, a business would want to keep separate records of the overhead of the business and the purchases from customers to be able to calculate an accurate revenue stream. We used foreign keys within Purchases to show what type of animal the customer purchased, which, in the future, could be broke down further to keep a data table that shows how long a customer as owned a pet. That data could potentially be used by the shop staff to have follow up calls planned to see if the customer needs anymore assistance. Lastly, we made the Loyalty Program entity. We decided to go somewhat basic with the attributes but this could be expanded in the future if the scenario arises. We have a foreign key within Purchases because the loyalty program depends on the amount/type of customer purchases. For the cardinality diagram at the bottom, we decided the relations were either many to many or one to many.
