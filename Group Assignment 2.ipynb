Our group has decided to break down our assignment into five different entities. We believe that this idea is a new and creative idea but has very similar ties to companies like GasBuddy.com. The center entity for our idea is Vehicles, with the other entities stemming from it. On one side, we have the owners of the vehicles along with the travel information per week. This entity will contain data on how many miles are driven per week for each driver. We also believe that there should be a distinction on the miles per gallon of the car and that data is shown by the Driving Setting (urban/highway). The data for the Owners will only have basic demographic data, like name and age. The vehicles entity contains quite a few data points. We believe all of the information is both useful and needed. The basic information of the make, model, and year of the vehicle are very important. We also believe that having the advertised fuel efficiency of the car would be a useful data point to have to be able to compare to the real-time and dynamic Travel Information. The other two entities are related to the gas stations that the owners are filling up at as well as the fuel types that are being used. This data can be applied to look at the driver habits of drivers either within a certain area of the data can be projected onto a national level project. Hopefully this information is useful and can be applied to the future projects.
